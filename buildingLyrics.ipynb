```python
from nltk.tag import pos_tag
stopwords = nltk.corpus.stopwords.words('english')
from nltk.corpus import wordnet as wn
wnl = WordNetLemmatizer()

tokens = []

for y in song_lyrics_list:
    per_song_tokens = []
    for x in y:
        x = str(x).lower()
        x = str(x).replace("in'","ing")
        x = str(x).replace("'","")
        word_tokens = wordpunct_tokenize(x)
        for y in word_tokens:
            per_song_tokens.append(y)
    tokens.append(per_song_tokens)
print(tokens)
```

```python
cleaned = []

for y in tokens:
    cleaned_per_song=[]
    for x in y:
        if len(x)>1:
            if x.isalpha():
                if x not in stopwords:
                    if x not in Most_common_words:
                        cleaned_per_song.append(x)
    cleaned.append(cleaned_per_song)

print(cleaned)
```

```python
lem_list = []

for y in cleaned:
    lem_list_per_song = []
    for x in y:
        if len(wn.synsets(x)) >= 1:
            if x[-3:] == "ing":
                word = wn.synsets(x)
                if '.v.' in str(word):
                    lem = wnl.lemmatize(x, pos='v')
                    lem_list_per_song.append(lem)
            else:
                word = wn.synsets(x)[0]
                tmp = word.pos()
                lem = wnl.lemmatize(x, pos=tmp)
                lem_list_per_song.append(lem)
    lem_list.append(lem_list_per_song)
```

```python
total_list = []

for y in lem_list:
    count = Counter(y).most_common(20)
    print(count)
    for x in count:
        total_list.append(x[0])
```
