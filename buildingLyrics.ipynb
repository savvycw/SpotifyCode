```python
TOKEN = os.getenv('GeniusID')
load_dotenv()

genius = Genius(TOKEN)
genius.verbose = False # Turn off status messages
genius.remove_section_headers = False

df = pd.read_json("C:\\Users\\savan\\OneDrive\\Documents\\Codeee\\SpotifyDF.json")
song_lyrics_list = []
for index, row in df.iterrows():
    x = row["Track Name"]
    y = row["Artist Name"]
    song = genius.search_song(x, y)
    lyrics = song.lyrics
    lyricLine = lyrics.splitlines()
    lyricList = []
    for x in lyricLine:
        if "[" not in x:
            lyricList.append(x)
    song_lyrics_list.append(lyricList)
#print(song_lyrics_list)
# res = re.findall('([a-zA-Z ]*)\d*.*', 'far12Embed"]')
# print(res)
words = []
for y in song_lyrics_list:
    tokens_per_song = []
    for x in y:
        sx = x.split(" ")
        for z in sx:
            z = z.replace(",","")
            if z[-3:] == "in'":
                z = z.replace("in'","ing")
            tokens_per_song.append(z)
    words.append(tokens_per_song)
print(words)
```

```python
from nltk.tag import pos_tag
stopwords = nltk.corpus.stopwords.words('english')
from nltk.corpus import wordnet as wn
wnl = WordNetLemmatizer()

tokens = []
for y in words:
    tokens_per_song = []
    for x in y:
        x = str(x).lower()
        word_tokens = word_tokenize(x)
        tokens_per_song.append(word_tokens)
    tokens.append(tokens_per_song)

stripped_by_song = []
for x in range(len(tokens)):
    stripped = []
    for y in range(len(tokens[x])):
        for word in tokens[x][y]:
            if word not in stopwords:
                if len(word) > 2:
                    if "'" not in word:
                        stripped.append(word)
    stripped_by_song.append(stripped)

origninal_set_lem = []
for w in stripped_by_song:
    song = []
    for x in w:
        list1 = []
        if len(wn.synsets(x)) >= 1:
            tmp = wn.synsets(x)[0].pos()
            list1 = [x,tmp]
            song.append(list1)
            #print(x, ":", tmp)
        else:
            list1 = [x, ""]
            song.append(list1)
    origninal_set_lem.append(song)


final_words_per_song = []
for x in origninal_set_lem:
    per_song = []
    for y in x:
        if y[1] == "":
            per_song.append(y[0])
        else:
            if y[0][-3:] == "ing":
                y[1] = 'v'
            lem = wnl.lemmatize(y[0], pos=y[1])
            per_song.append(lem)
    final_words_per_song.append(per_song)

combined_words = []
fully_combined = []
for x in final_words_per_song:
    counter = Counter(x)
    for x in counter:
        fully_combined.append(x)
total = Counter(fully_combined)
for x in counter.items():
    #print(x[0])
    if x[0] in Most_common_words:
        print(x[0])
        total.pop(x[0])
#print(total.most_common(20))


# #word vectors with training
# #catagories as the playlist name??
# remove most common 100 words
```
