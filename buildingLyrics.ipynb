```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
import time
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('maxent_treebank_pos_tagger')
```

```python
df = pd.read_json("C:\\Users\\savan\\OneDrive\\Documents\\Codeee\\SpotifyDF.json")

url_list = []
for index, row in df.iterrows():
    x = row["Track Name"]
    y = row["Artist Name"]
    x = re.sub(r'[^a-zA-Z0-9]', '', x)
    x = x.lower()
    y = re.sub(r'[^a-zA-Z0-9]', '', y)
    y = y.lower()
    if y[0:3] == 'the':
        y = y.replace('the','')
    url = "https://www.azlyrics.com/lyrics/" + y + "/" + x + ".html"
    url_list.append(url)


song_lyrics_list = []
for song in url_list:
    print(song)
    page = requests.get(song)

    soup = BeautifulSoup(page.content,"html.parser")
    ob_elements = soup.find_all('div',class_ = 'col-xs-12 col-lg-8 text-center')

    print(soup.prettify())

    new_element = str(ob_elements)
    splits = new_element.split("</div>")
    lyrics = splits[5]
    lyrics = lyrics.replace('<br/>','')
    #print(lyrics[1])
    lyricLine = lyrics.splitlines()
    lyricList = []
    for line in range(len(lyrics.splitlines())):
        if line > 4:
            lyricList.append(lyricLine[line])
            #should I delete repeated lines??
    song_lyrics_list.append(lyricList)
    time.sleep(30)
```

```python
from nltk.tag import pos_tag
stopwords = nltk.corpus.stopwords.words('english')
from nltk.corpus import wordnet as wn

tokens = []
for y in song_lyrics_list:
    tokens_per_song = []
    for x in y:
        x = str(x).lower()
        word_tokens = word_tokenize(x)
        tokens_per_song.append(word_tokens)
    tokens.append(tokens_per_song)

stripped_by_song = []
for x in range(len(tokens)):
    stripped = []
    for y in range(len(tokens[x])):
        for word in tokens[x][y]:
            if word not in stopwords:
                if len(word) > 2:
                    if "'" not in word:
                        stripped.append(word)
    stripped_by_song.append(stripped)

origninal_set_lem = []
for w in stripped_by_song:
    song = []
    for x in w:
        list1 = []
        if len(wn.synsets(x)) >= 1:
            tmp = wn.synsets(x)[0].pos()
            list1 = [x,tmp]
            song.append(list1)
            #print(x, ":", tmp)
        else:
            list1 = [x, ""]
            song.append(list1)
    origninal_set_lem.append(song)
#print(origninal_set_lem[1][1])

#print(new_combined_words)
final_words_per_song = []
for x in origninal_set_lem:
    per_song = []
    for y in x:
        if y[1] == "":
            per_song.append(y[0])
        else:
            if y[0][-3:] == "ing":
                y[1] = 'v'
            lem = wnl.lemmatize(y[0], pos=y[1])
            per_song.append(lem)
    final_words_per_song.append(per_song)
#print(final_words_per_song)

# print(stripped)
combined_words = []
for x in final_words_per_song:
    #print(x)
    counter = Counter(x)
    common = counter.most_common(10)
    for y in common:
        combined_words.append(y[0])

#print(combined_words)


counter2 = Counter(combined_words)
print(counter2)

# add ing to equal verb?
# #word vectors with training
# #catagories as the playlist name??
```
